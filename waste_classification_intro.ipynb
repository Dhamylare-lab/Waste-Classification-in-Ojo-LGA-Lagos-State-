## **Overview**
This project uses a Convolutional Neural Network (CNN) to classify different types of waste—**plastic, metal, paper, organic**, etc.—from images.  
The model is trained using TensorFlow and run on Google Colab, designed for environmental impact, especially in communities like **Ojo LGA, Lagos State**.

---

## **Dataset**
- A mini TrashNet-like dataset with folders:
  - `/plastic`, `/metal`, `/paper`, `/organic`, etc.
- Images resized to **150x150** for CNN input
- **80/20 split** for training and validation

---

## **Key Features**
- CNN model built with **TensorFlow/Keras**
- **Data augmentation** to improve generalization
- **Early stopping** and **model checkpointing**
- Custom prediction function
- Easy deployment on Google Colab with low compute
